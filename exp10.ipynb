```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.naive_bayes import GaussianNB
import ipywidgets as widgets
from IPython.display import display

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Initialize and train the model
model = GaussianNB()
model.fit(X, y)

#accuracy
from sklearn.metrics import accuracy_score
y_pred = model.predict(X)
accuracy = accuracy_score(y, y_pred)

# Define the feature names for user input
feature_names = iris.feature_names

# Create input widgets for each feature
feature_inputs = []
for name in feature_names:
    feature_inputs.append(widgets.FloatSlider(
        value=0.0,
        min=0.0,
        max=8.0 if 'length' in name else 5.0,
        step=0.1,
        description=name,
        continuous_update=False
    ))
#header and footer
header = widgets.HTML(value="<h1 style='text-decoration:underline'>Iris Flower Species Predictor</h1>")
footer = widgets.HTML(value="<p>&copy Gaussian Naive bayes by DIDAR, NISHANT and VISHAL</p>")

# Create a button to make predictions
predict_button = widgets.Button(description="Predict Species")

#create a label to show the prediction
prediction_label = widgets.Label("Predicted Species: ")

#accuracy label
accuracy_label = widgets.Label(f"With an accuracy of {accuracy*100:.2f}%")

def on_predict_button_click(b):
    # Gather inputs and predict species
    inputs = np.array([[slider.value for slider in feature_inputs]])
    prediction = model.predict(inputs)[0]
    species = iris.target_names[prediction]

    # Display the result
    prediction_label.value = f"Predicted Species: {species}"

# Bind the button click event to the function
predict_button.on_click(on_predict_button_click)

#verical alignment
main_layout=widgets.VBox([header,*feature_inputs,predict_button,prediction_label,accuracy_label,footer])

# Display the inputs and button
display(main_layout)


```
